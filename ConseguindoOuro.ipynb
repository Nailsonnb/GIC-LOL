```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn import preprocessing

#Carregamento dos dados
datatrain = pd.read_csv('D:/Nailson/5 periodo/GIC/league of References/agora vai/21-11/DiferencaOuro2.csv', 
                        sep=',')


# transformando dados num array
datatrain_array = datatrain.as_matrix()

```

```python
todos = []

for i in range(len(datatrain_array)):
    a = eval(datatrain_array[i][0])
    n1 = (len(a)//2)-1
    n2 = (n1+5)
    n3 = len(a)-5
    n4 = len(a)
    b = a[n1:n2]+a[n3:n4]
    todos.append(b)
```

```python

X_normalized = preprocessing.normalize(todos, norm='l1')
#X_scaled = preprocessing.scale(X_train)
#X_scaled

for i in range(len(X_normalized)):
    print(""+str(X_normalized[i][0])+","+str(X_normalized[i][1])+","+str(X_normalized[i][2])+","+str(X_normalized[i][3])+","+str(X_normalized[i][4])+","+str(X_normalized[i][5])+","+str(X_normalized[i][6])+","+str(X_normalized[i][7])+","+str(X_normalized[i][8])+","+str(X_normalized[i][9])+"")
```

```python
# Create a Pandas dataframe from some data.
df = pd.DataFrame({'Data': X_normalized})

# Create a Pandas Excel writer using XlsxWriter as the engine.
writer = pd.ExcelWriter('pandas_simple4.xlsx', engine='xlsxwriter')

# Convert the dataframe to an XlsxWriter Excel object.
df.to_excel(writer, sheet_name='Sheet1')

# Close the Pandas Excel writer and output the Excel file.
writer.save()
```
